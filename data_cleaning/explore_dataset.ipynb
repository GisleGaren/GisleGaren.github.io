```python
import pandas as pd

df_vehicle_crashes = pd.read_csv("../vehicle_crashes_info.zip")
```

```python
def summarize_columns(df):
    print(pd.DataFrame([
        (
            c,                               # Column name
            df[c].dtype,                     # Data type
            len(df[c].unique()),             # Number of unique values
            df[c].memory_usage(deep=True) // (1024**2)  # Memory usage in MB
        ) for c in df.columns
    ], columns=['name', 'dtype', 'unique', 'size (MB)']))

    print('Total size:', df.memory_usage(deep=True).sum() / 1024**2, 'MB')  # Total memory usage
    
summarize_columns(df_vehicle_crashes)
```

```python
# Convert CRASH DATE to datetime
df_vehicle_crashes['CRASH DATE'] = pd.to_datetime(df_vehicle_crashes['CRASH DATE'], format='%m/%d/%Y')

# Let's drop LOCATION because we already have LONTITUDE AND LATITUDE columns
df_vehicle_crashes.drop(columns=[
    'LOCATION', 'CROSS STREET NAME', 'OFF STREET NAME', 'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5',
    'VEHICLE TYPE CODE 4', 'VEHICLE TYPE CODE 5',
], inplace=True)

# Turn the street name to string
df_vehicle_crashes['ON STREET NAME'] = df_vehicle_crashes['ON STREET NAME'].astype('string')

# Let's make CONTRIBUTING FACTOR VEHICLE 1 to 5 categorical
df_vehicle_crashes['CONTRIBUTING FACTOR VEHICLE 1'] = df_vehicle_crashes['CONTRIBUTING FACTOR VEHICLE 1'].astype('category')
df_vehicle_crashes['CONTRIBUTING FACTOR VEHICLE 2'] = df_vehicle_crashes['CONTRIBUTING FACTOR VEHICLE 2'].astype('category')
df_vehicle_crashes['CONTRIBUTING FACTOR VEHICLE 3'] = df_vehicle_crashes['CONTRIBUTING FACTOR VEHICLE 3'].astype('category')

# Let's turn VEHICLE TYPE CODE 1 to 5 to categories
df_vehicle_crashes['VEHICLE TYPE CODE 1'] = df_vehicle_crashes['VEHICLE TYPE CODE 1'].astype('category')
df_vehicle_crashes['VEHICLE TYPE CODE 2'] = df_vehicle_crashes['VEHICLE TYPE CODE 2'].astype('category')
df_vehicle_crashes['VEHICLE TYPE CODE 3'] = df_vehicle_crashes['VEHICLE TYPE CODE 3'].astype('category')

# Now let's clean the borough column
df_vehicle_crashes['BOROUGH'] = df_vehicle_crashes['BOROUGH'].astype('category')

df_vehicle_crashes['ZIP CODE'] = pd.to_numeric(
    df_vehicle_crashes['ZIP CODE'], 
    errors='coerce'  # turn invalid strings into NaN
).astype('Int32')  # Use pandas nullable integer type

# Now let's look at how much data crashtime and crash date uses now
summarize_columns(df_vehicle_crashes)
```

```python
df_vehicle_crashes.to_parquet("vehicle_crashes_cleaned.parquet", index=False)
```
