```python
import pandas as pd

df_vehicle_crashes_vehicle = pd.read_csv("../vehicle_crashes_vehicle.zip")
```

```python
def summarize_columns(df):
    print(pd.DataFrame([
        (
            c,                               # Column name
            df[c].dtype,                     # Data type
            len(df[c].unique()),             # Number of unique values
            df[c].memory_usage(deep=True) // (1024**2)  # Memory usage in MB
        ) for c in df.columns
    ], columns=['name', 'dtype', 'unique', 'size (MB)']))

    print('Total size:', df.memory_usage(deep=True).sum() / 1024**2, 'MB')  # Total memory usage
```

```python
df_vehicle_crashes_vehicle.head(25)
```

```python
# Let's see the storage size of the columns
summarize_columns(df_vehicle_crashes_vehicle)
```

```python
# Let's drop the column VEHICLE_ID as it's not needed
df_vehicle_crashes_vehicle.drop(columns=['VEHICLE_ID','UNIQUE_ID','STATE_REGISTRATION','DRIVER_LICENSE_JURISDICTION','VEHICLE_DAMAGE_1', 'VEHICLE_DAMAGE_2', 'VEHICLE_DAMAGE_3', 'PUBLIC_PROPERTY_DAMAGE_TYPE'], inplace=True)
```

```python
# Let's see the storage size of the columns
summarize_columns(df_vehicle_crashes_vehicle)
```

```python
# Let's see the top 5 TRAVEL_DIRECTION 
df_vehicle_crashes_vehicle['TRAVEL_DIRECTION'].value_counts().head(5)
```

```python
# Convert CRASH DATE to datetime
df_vehicle_crashes_vehicle['CRASH_DATE'] = pd.to_datetime(df_vehicle_crashes_vehicle['CRASH_DATE'], format='%m/%d/%Y')

df_vehicle_crashes_vehicle['VEHICLE_TYPE'] = df_vehicle_crashes_vehicle['VEHICLE_TYPE'].astype('category')
df_vehicle_crashes_vehicle['VEHICLE_MAKE'] = df_vehicle_crashes_vehicle['VEHICLE_MAKE'].astype('category')
df_vehicle_crashes_vehicle['VEHICLE_MODEL'] = df_vehicle_crashes_vehicle['VEHICLE_MODEL'].astype('category')
df_vehicle_crashes_vehicle['TRAVEL_DIRECTION'] = df_vehicle_crashes_vehicle['TRAVEL_DIRECTION'].astype('category')
df_vehicle_crashes_vehicle['DRIVER_SEX'] = df_vehicle_crashes_vehicle['DRIVER_SEX'].astype('category')
df_vehicle_crashes_vehicle['DRIVER_LICENSE_STATUS'] = df_vehicle_crashes_vehicle['DRIVER_LICENSE_STATUS'].astype('category')
df_vehicle_crashes_vehicle['PRE_CRASH'] = df_vehicle_crashes_vehicle['PRE_CRASH'].astype('category')
df_vehicle_crashes_vehicle['POINT_OF_IMPACT'] = df_vehicle_crashes_vehicle['POINT_OF_IMPACT'].astype('category')
df_vehicle_crashes_vehicle['VEHICLE_DAMAGE'] = df_vehicle_crashes_vehicle['VEHICLE_DAMAGE'].astype('category')
df_vehicle_crashes_vehicle['PUBLIC_PROPERTY_DAMAGE'] = df_vehicle_crashes_vehicle['PUBLIC_PROPERTY_DAMAGE'].astype('category')
df_vehicle_crashes_vehicle['CONTRIBUTING_FACTOR_1'] = df_vehicle_crashes_vehicle['CONTRIBUTING_FACTOR_1'].astype('category')
df_vehicle_crashes_vehicle['CONTRIBUTING_FACTOR_2'] = df_vehicle_crashes_vehicle['CONTRIBUTING_FACTOR_2'].astype('category')
```

```python
# Let's see the storage size of the columns
summarize_columns(df_vehicle_crashes_vehicle)
```
