```python
import pandas as pd

df_vehicle_crashes_person = pd.read_csv("../vehicle_crashes_person.zip")
```

```python
df_vehicle_crashes_person.head(25)
```

```python
def summarize_columns(df):
    print(pd.DataFrame([
        (
            c,                               # Column name
            df[c].dtype,                     # Data type
            len(df[c].unique()),             # Number of unique values
            df[c].memory_usage(deep=True) // (1024**2)  # Memory usage in MB
        ) for c in df.columns
    ], columns=['name', 'dtype', 'unique', 'size (MB)']))

    print('Total size:', df.memory_usage(deep=True).sum() / 1024**2, 'MB')  # Total memory usage
```

```python
# Let's see the storage size of the columns
summarize_columns(df_vehicle_crashes_person)
```

```python
# Let's drop the column VEHICLE_ID as it's not needed
df_vehicle_crashes_person.drop(columns=['UNIQUE_ID','PERSON_ID','VEHICLE_ID','PED_ROLE','EMOTIONAL_STATUS','BODILY_INJURY','COMPLAINT'], inplace=True)
```

```python
# Let's see the storage size of the columns
summarize_columns(df_vehicle_crashes_person)
```

```python
# Convert CRASH DATE to datetime
df_vehicle_crashes_person['CRASH_DATE'] = pd.to_datetime(df_vehicle_crashes_person['CRASH_DATE'], format='%m/%d/%Y')
df_vehicle_crashes_person['PERSON_TYPE'] = df_vehicle_crashes_person['PERSON_TYPE'].astype('category')
df_vehicle_crashes_person['PERSON_INJURY'] = df_vehicle_crashes_person['PERSON_INJURY'].astype('category')
df_vehicle_crashes_person['EJECTION'] = df_vehicle_crashes_person['EJECTION'].astype('category')
df_vehicle_crashes_person['POSITION_IN_VEHICLE'] = df_vehicle_crashes_person['POSITION_IN_VEHICLE'].astype('category')
df_vehicle_crashes_person['SAFETY_EQUIPMENT'] = df_vehicle_crashes_person['SAFETY_EQUIPMENT'].astype('category')
df_vehicle_crashes_person['PED_ACTION'] = df_vehicle_crashes_person['PED_ACTION'].astype('category')
df_vehicle_crashes_person['PED_LOCATION'] = df_vehicle_crashes_person['PED_LOCATION'].astype('category')
df_vehicle_crashes_person['CONTRIBUTING_FACTOR_1'] = df_vehicle_crashes_person['CONTRIBUTING_FACTOR_1'].astype('category')
df_vehicle_crashes_person['CONTRIBUTING_FACTOR_2'] = df_vehicle_crashes_person['CONTRIBUTING_FACTOR_2'].astype('category')
df_vehicle_crashes_person['PERSON_SEX'] = df_vehicle_crashes_person['PERSON_SEX'].astype('category')
```

```python
# Let's see the storage size of the columns
summarize_columns(df_vehicle_crashes_person)
```

```python
df_vehicle_crashes_person.to_parquet("vehicle_crashes_person_cleaned.parquet", index=False)
```
